```python
%sql
create schema delta_catalog.raw
```

### creating a manged dekta tabke


```python
%sql
create table delta_catalog.raw.managed_my_table(
  id int,
  order_name string,
  amount double,
  prod_id int
)
using delta
```

```python
%sql
insert into delta_catalog.raw.managed_my_table
values (1, 'order1', 1001, 100),
       (2, 'order2', 2002, 200),
       (3, 'order3', 3003, 300)
```

```python
%sql
select  * from delta_catalog.raw.managed_my_table
```

```python
### external delta table
```

```python
%sql
create table delta_catalog.raw.ext_table(
  id int,
  order_name string,
  amount double,
  prod_id int
)
using delta
location 'abfss://raw@storagedeltalakepoorna.dfs.core.windows.net/ext_table'
```

```python
%sql
insert into delta_catalog.raw.ext_table
values (1, 'order1', 1001, 100),
       (2, 'order2', 2002, 200),
       (3, 'order3', 3003, 300)
```

```python
%sql
select * from delta.`abfss://raw@storagedeltalakepoorna.dfs.core.windows.net/ext_table`
```

CETAS

```python
%sql
create table delta_catalog.raw.ext_table_cetas
using delta
location 'abfss://raw@storagedeltalakepoorna.dfs.core.windows.net/ext_table_cetas'
as
select * from delta_catalog.raw.ext_table
```

### deep clone


```python
%sql
create table delta_catalog.raw.ext_table_deep
deep clone delta_catalog.raw.ext_table
```

###shallow clone

```python
%sql
create table delta_catalog.raw.ext_table_shallow
shallow clone delta_catalog.raw.managed_my_table
```
